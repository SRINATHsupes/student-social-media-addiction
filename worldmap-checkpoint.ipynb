```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from time import sleep

# Load data
data = pd.read_csv("Downloads/smb.zip")

# Count students per country
country_counts = data['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Count']
```

```python
geolocator = Nominatim(user_agent="geoapi")
country_counts['Latitude'] = None
country_counts['Longitude'] = None
for i, row in country_counts.iterrows():
    try:
        location = geolocator.geocode(row['Country'])
        if location:
            country_counts.at[i, 'Latitude'] = location.latitude
            country_counts.at[i, 'Longitude'] = location.longitude
        sleep(1)  # pause to avoid getting blocked
    except:
        print(f"Could not locate: {row['Country']}")
```

```python
world_map = folium.Map(location=[20, 0], zoom_start=2)

# Add markers for each country
for _, row in country_counts.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=row['Count'] * 0.5,
            popup=f"{row['Country']}: {row['Count']} students",
            color='blue',
            fill=True,
            fill_opacity=0.7
        ).add_to(world_map)
```
